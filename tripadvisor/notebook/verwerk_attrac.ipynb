```python
import folium
from folium.plugins import MarkerCluster, FastMarkerCluster
import geopandas as gpd
import contextily as ctx
import matplotlib.pyplot as plt

%matplotlib inline
CRS = '+proj=sterea +lat_0=52.15616055555555 +lon_0=5.38763888888889 +k=0.9999079 +x_0=155000 +y_0=463000 +ellps=bessel +units=m +towgs84=565.2369,50.0087,465.658,-0.406857330322398,0.350732676542563,-1.8703473836068,4.0812 +no_defs no_defs'
```

```python
gdf = gpd.GeoDataFrame(df_mra, geometry=gpd.points_from_xy(df_mra.lat, df_mra.lon), crs=CRS)
```

```python
cb =  """
function (row) {
    var icon0 = L.AwesomeMarkers.icon({
        prefix: 'fa',
        icon: 'coffee',
        markerColor: 'blue',
        iconColor: 'white'
    });
    var icon1 = L.AwesomeMarkers.icon({
        prefix: 'fa',
        icon: 'blind',
        markerColor: 'red',
        iconColor: 'white'
    });
    var marker = L.marker(new L.LatLng(row[0], row[1]));
    if (row[4] == 1){
        marker.setIcon(icon1);
    } else {
        marker.setIcon(icon0)
    }
    marker.bindPopup('<h3>'+row[2]+'</h3>'+'</br>'+row[3]);
    return marker;
};
"""
```

```python
m = folium.Map(location=[52.38, 4.9], zoom_start=11)

m.add_child(
    FastMarkerCluster(
        df_mra[['lat', 'lon', 'titel', 'adres', 'categorie_Tours']].values.tolist(), 
        callback=cb
    )
)

m.save('kaartje.html')
```
